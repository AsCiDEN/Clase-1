### **Clases de python**

Sabemos que python es un lenguage de programación orientado a objetos, es decir, esta enfocado en la organización y manipulación de datos. Es por esto que los objetos en Python esán estructurados como clase. Uno de los problemas de las funciones de python es que no almacenan, solo operan: entregamos un input y obtenemos un output. A diferencia de las funciones, en las clases podemos almacenar información en forma de atributos o propiedades. Adicionalmente, las clases de Python poseen funciones internas, las cuales pueden recibir como input tanto los atributos internos de la clase así como inputs externos. Más adelante revisaremos la creación de una clase.

En el mejor de los casos (como pasa con librerías conocidas como Numpy y Matplotlib) la librería esta bien documentada en una página web, por lo que podemos aprender a trabajar con el objeto mediante una interfaz amigable. #### Creando una clase

Ahora repasaremos como crear una clase. Como ejemplo, necesitamos crear la clase planeta:

````python
import numpy as np
class planeta:
    def __init__(self, mass, x, y, z, vx, vy, vz, name):
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.mass = mass
        self.name = name

    def print(self):
        print('Planet name: ', self.name)
        print('mass: ', self.mass)
        print('position: ', self.x, self.y, self.z)
        print('velocity: ', np.sqrt(self.vx**2 + self.vy**2 + self.vz**2))
        return

    def update_force(self, fx, fy, fz, dt):
        ax = fx/self.mass
        ay = fy/self.mass
        az = fz/self.mass

        # actualizamos posiciones
        self.x += self.vx*dt + 0.5*ax*dt**2
        self.y += self.vy*dt + 0.5*ay*dt**2
        self.z += self.vz*dt + 0.5*az*dt**2

        # actualizamos velocidades
        self.vx += ax*dt
        self.vy += ay*dt
        self.vz += az*dt
        return
````

Notamos que con el método init definimos los atributos principales. Lo interesante de este método es que nos permite inicializar el objeto como si fuese una función:

````python
PlanetaTierra = planeta(1, 0, 0, 0, 0.5, 0.5, 0, 'Tierra')
PlanetaTierra.print()

print('')
print('--------------------')
print('')

#Creamos una fuerza ficticia para actualizar valores
Fx, Fy, Fz = 1, -1, 0
PlanetaTierra.update_force(Fx, Fy, Fz, 1)
PlanetaTierra.print()
````

Podemos modificar y agragar atributos

````python
PlanetaTierra.mass = 2
PlanetaTierra.print()

print('')
print('---------')
print('')

PlanetaTierra.radius = 1.5
print(PlanetaTierra.radius)
````

## Herramientas fundamentales: **Numpy**
---

Numpy es con seguridad la librería mas utilizada en ciencia de datos. Nos permite operar con una mayor cantidad de datos y es mas eficiente que las listas de python. Además, su basta cantidad de clases y funciones estan enfocadas principalmente al uso científico. Podemos hacer con facilidad operaciones vectoriales y matriciales, entre otras cosas. Es por esto que es la base de un gran número de librerías en python.

numpy.jpg




````python
import numpy as np # Importamos la librería
````

````python
###--- Arrays(vectores) ---###

# Se crean a partir de una lista de python

array_1 = np.array(['a', 'b', 'c', 'd', 'e'])
print(array_1)

lista = [1., 2., 3., 4., 5., 6.]
array_2 = np.array(lista)
print(array_2)
````

````python
### Operaciones básicas

array1 = np.array([0, 1, 2, 3, 4, 5, 6, 13, 7, 1, 2], dtype = float) # nota: con dtype estamos especificando que queremos setear el tipo de elemento del array
array2 = np.linspace(1, 5, 11) # funcion linspace de numpy: generamos un vector de 10 elementos equiespaciados, en el rango (1, 5)

# largo
print('Largo (arr1): ', np.size(array1)) # ¿que diferencia hay con len(array1)?

# suma
print('Suma: ', array1 + array2)

# multiplicacion
print('Multiplicacion: ', array1*array2)
````

````python
### Algebra lineal

# Producto interno
print('Producto interno', np.dot(array1, array2))

# Norma
print('Norma: ', np.linalg.norm(array1))

# Igual podemos calcular la norma sin utilizar dicha función haciendo
print('Norma (v2): ', np.sqrt(np.sum(array1**2)) )
````

````python
###--- Matrices ---###
fila1 = [1, 0, 3]
fila2 = [0, 3, 9]
fila3 = [1, 4, 5]
matrix1 = np.array([fila1, fila2, fila3])
print(matrix1)

matrix2 = np.array([[0.1, 0.3, 0.],
                    [1.1, 1.2, 1.4],
                    [0., 2.2, 2.3]])
print(matrix2)
````

````python
# Operaciones matriciales

# Producto punto
print('Producto punto: ')
print(np.dot(matrix1, matrix2))
print('')
print('')

# Producto interno
print('Producto interno: ')
print(np.inner(matrix1, matrix2))
print('')
print('')

# Determinante de una matriz
print('Determinante: ')
print(np.linalg.det(matrix1))
print('')
print('')

# Inversa de una matriz
print('Inversa: ')
print(np.linalg.inv(matrix2))
print(matrix2.dot(np.linalg.inv(matrix2)))
print('')
print('')

# Transpuesta de una matriz
print('Transpuesta de una matriz: ')
print(matrix1)
print(matrix1.T)
````

## Partimos haciendo un gráfico simple de varias funciones

````python
# Función lineal en x
def fn_lineal(x, m, n):
    return m*x + n

x = np.linspace(0, 2*np.pi, 50) # Vector equiespaciado

# Función lineal
m_1 = (1.1 - (-1))/(2*np.pi - 0)
n_1 = -1
y_1 = fn_lineal(x, m = m_1, n = n_1)

# Función coseno
fn_cos = np.cos(x)

# Función seno
fn_sin = np.sin(x)
````

````python
# Gráfico simple
plt.plot(fn_cos)
plt.plot(fn_sin)
plt.plot(y_1)
plt.show()
````

La función plt.plot recibe no solamente los datos a graficas, sino que también un set de parámetros los cuales nos permitirán controlar la estética del gráfico. Vemos que podemos fijar el color de la línea (color),estilo de línea (ls), tipo de marcador (marker), entre otros.

````python
# Mejoramos: Agregamos linestyle (ls), marker y color
plt.plot(x, fn_sin, color = 'C8', ls = '', marker = '.')
plt.plot(x, fn_cos, color = 'C9', ls = 'dashed', marker = '.')
plt.plot(x, y_1, color = 'mediumorchid')
plt.show()
````

````python
# Mejoramos Agregamos leyenda, límites, grilla

# Plot de datos
plt.plot(x, fn_sin, color = 'C8', ls = '', marker = '.', label = 'sin(x)')
plt.plot(x, fn_cos, color = 'C9', ls = 'solid', marker = 'o', mfc=(0.75, 0, 0, 0.8), mec='white', label = 'cos(x)')
plt.plot(x, y_1, color = 'mediumorchid', label = 'lineal')

# Configuraciones del plot
plt.xlim(np.min(x), np.max(x))  # limitamos x e y
plt.ylim(-1, 1.1)
plt.grid(True, ls = '--') # agregamos una grilla

plt.xlabel('x', fontsize=15) # texto mostrado en eje x
plt.ylabel('y = f(x)', fontsize=15) # texto mostrado en eje x
plt.title('sin(x), cos(x), L = m*x + n', fontsize=15) # texto mostrado en titulo
plt.legend(loc = 'upper center') # ubicamos la leyenda en el gráfico
plt.show()
````

## Trabajamos con un dataset

Ahora trabajaremos con datos de una tabla. Usaremos datos de la librería seaborn

````python
import seaborn as sns
import pandas as pd

Data = sns.load_dataset("tips")  # Datos de propina
print(Data.columns)              # Queremos ver columnas del dataset
Data.head()
````

Queremos separar los datos entre fumadores y no fumadores, por lo que hacemos un query en el set de datos

````python
# Primero revisamos los valores para hacer el query
Data['smoker'].unique()
````

````python
# Separamos entre fumadores y no fumadores
Data_smoker = Data.query('smoker == "Yes"')
Data_no_smoker = Data.query('smoker == "No"')
````

````python
# Graficamos las columnas total_bill y tip

plt.plot(Data_no_smoker['total_bill'], Data_no_smoker['tip'], ls = '', marker = 'o', mfc='orange', mec='black', label = 'No', alpha = 0.5)
plt.plot(Data_smoker['total_bill'], Data_smoker['tip'], ls = '', marker = 'o', mfc='green', mec='black', label = 'Yes', alpha = 0.5)
plt.legend(loc = 'upper left', title = 'Smoker')
plt.xlabel('Total boleta [$]')
plt.ylabel('Propina [$]')
plt.show()
plt.close()
````

Ahora graficaremos la asistencia total por día, por lo que haremos un histograma. Notemos que los datos de la columna "day" no son numéricos, por lo que deberemos "mapearlos" para poder hacer el histograma

````python
# Crearemos diccionarios para pasar de día a número y viceversa
days_dict = {}
invert_dict = {}
count = 1

for i in Data['day'].unique():
    days_dict[i] = count
    invert_dict[count] = i
    count = count + 1

print(days_dict)
print(invert_dict)
````

````python
# Agragamos una columna al dataset con los datos mapeados

days_to_num = Data['day'].apply(lambda x: days_dict[x]) # Aplicamos función a los datos de la columna day

Data['day_num'] = pd.to_numeric(days_to_num) # agregamos la columna como numérica
Data.head()
````

````python
Data['day_num'].hist(bins = 6, edgecolor='black', range = [-0.5, 5.5], density=True) # Histograma de los datos
plt.grid(True)
plt.xlabel('Dia')
plt.ylabel('Asistencia (%)')
plt. title('Histograma')
plt.xticks(list(invert_dict.keys()), list(days_dict.keys()))
plt.show()
````

Finalmente, generamos un gráfico doble. Notemos que usamos la funcion plt.subplot(m, n, k):

 - m: indica el número de filas
 - n: indica el número de columnas
 - k: dado que tenemos una grilla de $m\times n$, k nos indica la posición del gráfico en la misma, siguiendo el orden izquierda-derecha, arriba-abajo.

````python
fig = plt.figure(figsize=(12,5))
fig.suptitle('tips Dataset', fontsize=25)

plt.subplot(1, 2, 1) # Graficamos en la posición (0, 0) de la grilla
plt.plot(Data_no_smoker['total_bill'], Data_no_smoker['tip'], ls = '', marker = 'o', mfc='orange', mec='black', label = 'No', alpha = 0.5)
plt.plot(Data_smoker['total_bill'], Data_smoker['tip'], ls = '', marker = 'o', mfc='green', mec='black', label = 'Yes', alpha = 0.5)
plt.legend(loc = 'upper left', title = 'Smoker')
plt.xlabel('Total boleta [$]')
plt.ylabel('Propina [$]')


plt.subplot(1, 2, 2) # Graficamos en la posición (1, 0) de la grilla
plt.grid(True, zorder=0)
Data['day_num'].hist(bins = 6, edgecolor='black', range = [-0.5, 5.5], density=True)
plt.xlabel('Dia')
plt.ylabel('Asistencia (%)')
plt.title('Histograma')
plt.xticks(list(invert_dict.keys()), list(days_dict.keys()))

plt.savefig('plot.png')
plt.show()
plt.close()
````
